```python
from mlxtend.frequent_patterns import apriori
```

```python
dataset=[['Drink','Nuts','Diaper'],
        ['Drink','Coffee','Diaper'],
        ['Drink','Diaper','Eggs'],
        ['Nuts','Eggs','Milk'],
        ['Nuts','Coffee','Diaper','Eggs','Milk']]
```

```python
dataset
```

```python
import pandas as pd

from  mlxtend.preprocessing import TransactionEncoder

TranEncode = TransactionEncoder()

te_ary = TranEncode.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=TranEncode.columns_)
```

```python
df
```

let us return items and itemsets with 

```python
apriori(df, min_support=0.6)
```

```python
apriori(df, min_support=0.6, use_colnames = True)
```

```python
frequnt_itemsets = apriori(df,min_support=0.2,use_colnames=True)
frequnt_itemsets['length']= frequnt_itemsets['itemsets'].apply(lambda x : len(x))
frequnt_itemsets
```

```python
frequnt_itemsets[ (frequnt_itemsets['length'] == 5)&

                  (frequnt_itemsets['support'] >= 0.2)]
```

```python
frequnt_itemsets[ frequnt_itemsets['itemsets'] == {'Diaper', 'Drink'} ]
```

```python
x=frequnt_itemsets[ frequnt_itemsets['itemsets'] == {'Diaper', 'Drink'} ]
print(x['support'])
```

```python
from mlxtend.frequent_patterns import association_rules
association_rules(frequnt_itemsets, metric="confidence", min_threshold=0.7)
```
